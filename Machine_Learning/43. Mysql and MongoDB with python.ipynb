```python
create_table_query = """
CREATE TABLE IF NOT EXISTS computer (
    Company VARCHAR(255),
    TypeName VARCHAR(255),
    Inches FLOAT,
    ScreenResolution VARCHAR(255),
    Cpu VARCHAR(255),
    Ram VARCHAR(20),
    Memory VARCHAR(255),
    Gpu VARCHAR(255),
    OpSys VARCHAR(255),
    Weight VARCHAR(20),
    Price FLOAT
)
"""
```

```python
insert_query = """
INSERT INTO computer (Company, TypeName, Inches, ScreenResolution, Cpu, Ram, Memory, Gpu, OpSys, Weight, Price)
VALUES ( %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s)
ON DUPLICATE KEY UPDATE 
    Company=VALUES(Company), TypeName=VALUES(TypeName), Inches=VALUES(Inches),
    ScreenResolution=VALUES(ScreenResolution), Cpu=VALUES(Cpu), Ram=VALUES(Ram),
    Memory=VALUES(Memory), Gpu=VALUES(Gpu), OpSys=VALUES(OpSys), Weight=VALUES(Weight), Price=VALUES(Price)
"""
```

```python
data = df.iloc[:, 2:]
```

```python
data.head()
```

```python
cursor.execute(create_table_query)
```

```python
data_tuples= [tuple(i) for i in data.values]
```

```python
cursor.executemany(insert_query, data_tuples)
```
